```python
headers = range(0,25)
keys = map(lambda x: str(x), headers)
data = pd.read_csv('/home/cristiano/Documentos/Programming/Python/Python-Daily/scikit-learn/database_2.csv', sep=",", header = None)
X = data.ix[:,:data.shape[1]-2]
y = data.ix[:,data.shape[1]-1]
print X
print y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
regr = RandomForestRegressor(n_estimators=500,random_state=0)
regr.fit(X_train, y_train)
```

```python
print regr.score(X_test, y_test)
print regr.score(X_train, y_train)
```

```python
print regr.predict([[1,35,7,3,22,4,5,9,13,0,29]])
```

```python
result_list=[]
for elem in prediction_list:
    result_list.append(elem.tolist()[0])
print result_list
print len(result_list)
print len(lista)
```
